```python
import googlemaps

# Initialize the Google Maps client
gmaps = googlemaps.Client(key=API_KEY)

# Define the location (latitude and longitude) you want to search near
latitude = 30.6565  
longitude = 76.7387 
location = (latitude, longitude)

# Define the search radius (in meters) - e.g., 1000 meters (1 km)
radius = 1000

# Define the type of place you are searching for (e.g., 'restaurant', 'park', 'coffee_shop')
place_type = 'restaurant'

try:
    # Perform Nearby Search
    places_result = gmaps.places_nearby(location=location, radius=radius, type=place_type)

    if places_result['status'] == 'OK':
        print(f"Found restaurants near ({latitude}, {longitude}):")
        for place in places_result['results']:
            print(f"- {place['name']} ({place.get('vicinity', 'Address not available')})") # .get() handles missing 'vicinity'

        # You can access more details about each place using Place Details API
        # For example, to get details of the first restaurant:
        if places_result['results']:
            first_place_id = places_result['results'][0]['place_id']
            place_details_result = gmaps.place(place_id=first_place_id)
            if place_details_result['status'] == 'OK':
                print("\nDetails of the first restaurant:")
                print(f"  Name: {place_details_result['result']['name']}")
                print(f"  Address: {place_details_result['result'].get('formatted_address', 'Not available')}") # .get() for safety
                print(f"  Rating: {place_details_result['result'].get('rating', 'Not rated')}") # .get() for safety
                # ... access other details as needed (opening hours, reviews, etc.)
            else:
                print(f"Error getting details for place ID {first_place_id}: {place_details_result['status']}")

    else:
        print(f"Error during Places Nearby Search: {places_result['status']}")
        print(f"Error message: {places_result.get('error_message', 'No error message provided')}") # .get() for safety

except googlemaps.exceptions.ApiError as e:
    print(f"Google Maps API Error: {e}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

```python
import googlemaps

# Initialize the Google Maps client
gmaps = googlemaps.Client(key=API_KEY)

# Define origin and destination addresses or coordinates
origin = "Plakha University Mohali"  # Example origin address
destination = "IISER Mohali" # Example destination address

# You can also use coordinates (latitude, longitude) instead of addresses:
# origin = (40.785091, -73.968285) # Central Park coordinates
# destination = (40.758896, -73.985130) # Times Square coordinates

try:
    # Get walking directions
    directions_result = gmaps.directions(origin,
                                        destination,
                                        mode="walking")

    if directions_result: # Check if directions_result is not empty (API returned a route)
        route = directions_result[0]['legs'][0] # Get the first (and usually only) route leg

        # Extract distance and duration
        distance = route['distance']['text'] # e.g., "1.9 km" or "1.2 mi"
        duration = route['duration']['text'] # e.g., "25 mins"

        print(f"Walking directions from '{origin}' to '{destination}':")
        print(f"Distance: {distance}")
        print(f"Duration: {duration}")

        print("\nSteps:")
        for step in route['steps']:
            instruction = step['html_instructions'] # HTML formatted instructions
            distance_step = step['distance']['text']
            print(f"- {instruction} ({distance_step})")

    else:
        print(f"No walking route found from '{origin}' to '{destination}'.")

except googlemaps.exceptions.ApiError as e:
    print(f"Google Maps Directions API Error: {e}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

```python
import googlemaps
import folium


# Initialize the Google Maps client
gmaps = googlemaps.Client(key=API_KEY)

# Chandigarh City Center Coordinates (approximate)
chandigarh_latitude = 30.7333
chandigarh_longitude = 76.7794
chandigarh_location = (chandigarh_latitude, chandigarh_longitude)

# Search radius (in meters) - adjust as needed to cover Chandigarh city
search_radius = 5000  # 5 kilometers - should cover a good portion of Chandigarh

# Place type to search for
place_type = 'restaurant'

try:
    # Perform Nearby Search for restaurants in Chandigarh
    places_result = gmaps.places_nearby(location=chandigarh_location, radius=search_radius, type=place_type)

    if places_result['status'] == 'OK':
        restaurants = places_result['results']
        print(f"Found {len(restaurants)} restaurants in Chandigarh.")

        # Create a folium map centered on Chandigarh
        chandigarh_map = folium.Map(location=chandigarh_location, zoom_start=13) # zoom_start adjust zoom level

        # Add markers for each restaurant to the map
        for restaurant in restaurants:
            restaurant_lat = restaurant['geometry']['location']['lat']
            restaurant_lng = restaurant['geometry']['location']['lng']
            restaurant_name = restaurant['name']

            folium.Marker(
                location=[restaurant_lat, restaurant_lng],
                popup=restaurant_name,
                icon=folium.Icon(color='blue', icon='info-sign') # You can customize marker icon
            ).add_to(chandigarh_map)

        chandigarh_map


    else:
        print(f"Error during Places Nearby Search: {places_result['status']}")
        print(f"Error message: {places_result.get('error_message', 'No error message provided')}")

except googlemaps.exceptions.ApiError as e:
    print(f"Google Maps API Error: {e}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
chandigarh_map
```
